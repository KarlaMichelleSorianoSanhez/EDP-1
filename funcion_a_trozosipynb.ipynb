# Serie de Fourier de una función a trozos

La función esta definida por:


$$
f(x) =
\begin{cases}
0 & \text{si } -2 \leq x < 0 \\
2-x & \text{si } 0 \leq x \leq 2
\end{cases}
$$


Al calcular los coeficientes de la serie de Fourier, obtuvimos lo siguiente:
$$a_0=1$$

$$a_n=\frac{2(1-(-1)^n)}{(n\pi)^2}$$

$$b_n=\frac{-2}{n\pi}$$

Por lo tanto la serie que le corresponde a la funciónes:


$$g(x)=\frac{1}{2}+\frac{2}{\pi^2}\sum_{n=1}^{\infty}(\frac{2(1-(-1)^n)}{(n\pi)^2} \cos(\frac{n\pi(x)}{2})- \frac{-2}{n\pi}  \sin(\frac{n\pi(x)}{2})).$$

A continuación graficaremos nuestra función y nuestra serie de Fourier

# Define una función por partes



Define una función llamada *recta $(x)$*, que toma un $x$ como entrada y devuelve $y$ con los mismos valores, es decir, representa una función identidad $u(x,0)=x$

```python

# Definimos la función a trozos
def recta(x):
    """
    Función definida a trozos:
    - f(x) = 0 para -2 ≤ x < 0
    - f(x) = 2 - x para 0 ≤ x ≤ 2
    """
    y = zeros(len(x))  # Inicializa el vector de salida con ceros
    for i in range(len(x)):
        if x[i] < 0:
            y[i] = 0  # Primer tramo
        else:
            y[i] = 2 - x[i]  # Segundo tramo
    return y
```

## Graficar condición inical

Genera la gráfica de la condición inicial  $u(x,0)=x$ usando **linspace** para definir el dominio de la función y **plot** para dibujarla

```python
x = linspace(-2,2,1000)
f = recta(x)
plot(x,f,label='$f(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Gráfica de nuestra función definida a trozos")
xlim([-2.01,2.01]),ylim([-0.1,2.01]),grid(),legend()
show()
```

# Aproximación de la serie de fourier

¿Qué hace esta función?
Aproxima la función $f(x)=x$ usando una serie de fourier con $N$ términos. Cuantos más terminos tenga, más se parecera a la función original

Definimos nuestra serie de Fourier y la graficamos

```python
def u_0(x,N=20): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += ((1-(-1)**n)/n**2)*cos(n*pi*x/2)-pi/n*sin(n*pi*x/2)
    return (2/pi**2)*f+1/2
```

```python
f_n = u_0(x)
plot(x,f_n,label='$f(x)$ con $N=20$')
plot(x,f,label='$g(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Función definida a trozos y Serie de Fourier de la misma")
xlim([-2.01,2.1]),ylim([-2.01,2.2]),grid(),legend()
show()
```

Podemos visualizar la aproximación para diferentes cantidades de armónicos.

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2,2]),ylim([-2,2.2]),grid(),legend()
show()
```
